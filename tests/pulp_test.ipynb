```python
import pandas as pd
from pulp import LpMaximize, LpProblem, lpSum, LpVariable, value

data = pd.read_csv(r"C:\Users\simon\6_Semester\kip\KIP_texprompter\texprompter\data\optimization_pipeline_test_easy.csv")
model = LpProblem(name="optimization_problem", sense=LpMaximize)
product_ids = data["Product_ID"]
x = LpVariable.dicts("Product_ID", product_ids, lowBound=0, upBound=None, cat='Integer')
model += lpSum([data.loc[i, "Profit_Per_Unit"] * x[product_ids[i]] for i in range(len(product_ids))])
model += lpSum([data.loc[i, "Machine_A_Hours_Req"] * x[product_ids[i]] for i in range(len(product_ids))]) <= 10
model += lpSum([data.loc[i, "Machine_B_Hours_Req"] * x[product_ids[i]] for i in range(len(product_ids))]) <= 10
model += lpSum([data.loc[i, "Labor_Hours_Req"] * x[product_ids[i]] for i in range(len(product_ids))]) <= 10
model += lpSum([data.loc[i, "Raw_Material_Units_Req"] * x[product_ids[i]] for i in range(len(product_ids))]) <= 10
status = model.solve()
print(f"Objective value: {value(model.objective)}")

for var in x.values():
    print(f"{var.name}: {var.value()}")
```
